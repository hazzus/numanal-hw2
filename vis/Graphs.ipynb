```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('../task2/iterations_dominated') as f:
    jacobi_dominated = list(map(float, f.read().split()))
len(jacobi_dominated)
```

```python
with open('../task3/iterations_dominated') as f:
    seidel_dominated = list(map(float, f.read().split()))
len(seidel_dominated)
```

```python
with open('../task4/iterations_dominated') as f:
    seidel2_dominated = list(map(float, f.read().split()))
len(seidel2_dominated)
```

```python
with open('../task5/iterations_dominated') as f:
    descent_dominated = list(map(float, f.read().split()))
len(descent_dominated)
```

```python
list(map(len, [jacobi_dominated, seidel_dominated, seidel2_dominated, descent_dominated]))
```

```python
minlen = min(list(map(len, [jacobi_dominated, seidel_dominated, seidel2_dominated, descent_dominated])))
minlen
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 17), )

axs[0].set_title('Сравнение методов', fontsize=20)
axs[0].set_xlabel('Количество итераций', fontsize=15, labelpad=10)
axs[0].set_ylabel('$|Ax - b|$', rotation=0, fontsize=15, labelpad=30)
axs[0].grid(True)
axs[0].plot(range(minlen), seidel_dominated[:minlen], label='Зайдель', linewidth=3)
axs[0].plot(range(minlen), seidel2_dominated[:minlen], label='Зайдель (рел.)', linewidth=3)
axs[0].plot(range(minlen), descent_dominated[:minlen], label='Спуск', linewidth=3)
axs[0].plot(range(minlen), jacobi_dominated[:minlen], label='Якоби', linewidth=3)
axs[0].legend()

axs[1].set_title('Сравнение методов (Последние итерации)', fontsize=20)
axs[1].set_xlabel('Количество итераций', fontsize=15, labelpad=10)
axs[1].set_ylabel('$|Ax - b|$', rotation=0, fontsize=15, labelpad=30)
axs[1].grid(True)
axs[1].plot(range(minlen), seidel_dominated[:minlen], label='Зайдель', linewidth=3)
axs[1].plot(range(minlen), seidel2_dominated[:minlen], label='Зайдель (рел.)', linewidth=3)
axs[1].plot(range(minlen), descent_dominated[:minlen], label='Спуск', linewidth=3)
axs[1].legend()
axs[1].set_xlim(7, float(minlen) - 1 + 0.05)
axs[1].set_xticks(range(7, 11))
axs[1].set_ylim(0, 0.2)
axs[1].set_yticks(np.arange(0, 0.21, 0.05))

plt.savefig('./graphs/dominatedNew.png', dpi=300)
```

```python
with open('../task3/iterations_random') as f:
    seidel_random = list(map(float, f.read().split()))
len(seidel_random)
```

```python
with open('../task5/iterations_random') as f:
    descent_random = list(map(float, f.read().split()))
len(descent_random)
```

```python
minlen_random = min(list(map(len, [seidel_random, descent_random])))
```

```python
fig, axs = plt.subplots(1, 1, figsize=(13, 7))

plt.plot(range(minlen_random)[1:], seidel_random[1:minlen_random], label='Зайдель', linewidth=3)
plt.plot(range(minlen_random)[1:], descent_random[1:minlen_random], label='Спуск', linewidth=3)
plt.ylabel('$|Ax - b|$', rotation=0, fontsize=15, labelpad=30)
plt.xlabel('Итерации', rotation=0, fontsize=15, labelpad=5)
plt.title('Сравнение методов', fontsize=20)
plt.legend()
plt.grid(True)

plt.savefig('./graphs/randomNew.png', dpi=300)
```
