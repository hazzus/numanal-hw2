```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
list(map(len, [jacobi_dominated, seidel_dominated, seidel2_dominated, descent_dominated]))
```

```python
minlen = min(list(map(len, [jacobi_dominated, seidel_dominated, seidel2_dominated, descent_dominated])))
minlen
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 17), )

axs[0].set_title(' ', fontsize=20)
axs[0].set_xlabel(' ', fontsize=15, labelpad=10)
axs[0].set_ylabel('$|Ax - b|$', rotation=0, fontsize=15, labelpad=30)
axs[0].grid(True)
axs[0].plot(range(minlen), seidel_dominated[:minlen], label='Seidel', linewidth=3)
axs[0].plot(range(minlen), seidel2_dominated[:minlen], label='Seidel (relaxed)', linewidth=3)
axs[0].plot(range(minlen), descent_dominated[:minlen], label='Descent', linewidth=3)
axs[0].plot(range(minlen), jacobi_dominated[:minlen], label='Jacobi', linewidth=3)
axs[0].legend()

axs[1].set_title('  ( )', fontsize=20)
axs[1].set_xlabel(' ', fontsize=15, labelpad=10)
axs[1].set_ylabel('$|Ax - b|$', rotation=0, fontsize=15, labelpad=30)
axs[1].grid(True)
axs[1].plot(range(minlen), seidel_dominated[:minlen], label='Seidel', linewidth=3)
axs[1].plot(range(minlen), seidel2_dominated[:minlen], label='Seidel (relaxed)', linewidth=3)
axs[1].plot(range(minlen), descent_dominated[:minlen], label='Descent', linewidth=3)
axs[1].legend()
axs[1].set_xlim(6, 10.05)
axs[1].set_xticks(range(6, 11))
axs[1].set_ylim(0, 0.2)
axs[1].set_yticks(np.arange(0, 0.21, 0.05))

plt.savefig('./graphs/dominatedNew.png', dpi=300)
```
